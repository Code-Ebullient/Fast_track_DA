```python
# Step 1 - Importing all the necessary libraries 

import pandas as pd
import seaborn as sns
import numpy as np

# Reading the datasets
rd = pd.read_csv('/Users/concept/Downloads/accenture/Reactions_file.csv')
cd = pd.read_csv('/Users/concept/Downloads/accenture/Content_file.csv')
rtd = pd.read_csv('/Users/concept/Downloads/accenture/ReactionTypes_file.csv')
```

```python
# Displaying the heads of all the datasets
display('rd.head()', 'cd.head()', 'rtd.head()')
```

```python
rd.head()
```

```python
# Merging two datasets (Reactions and Content) joining them on Content ID column
mg = pd.merge(rd, cd, on='Content ID', how='right')
```

```python
cd.head()
```

```python
# The new merged table
mg.head()
```

```python
# Checking for null values on the merged table
mg.isnull().any()
```

```python
# Taking a closer look at Type column
mg[mg['Type'].isnull()].head(10)
```

```python
# Filling the null values on Type column with a string
mg['Type'] = mg['Type'].fillna('no record')
```

```python
# Rechecking for null values
mg.isnull().any()
```

```python
# Merging the final table
merged = pd.merge(mg, rtd, on='Type', how='left')
```

```python
merged.head()
```

```python
# Checking for null values ones again
merged.isnull().any()
```

```python
# So may null value on Score column
merged[merged['Score'].isnull()].head(20)
```

```python
# Dropping the null values rows
merged.dropna(inplace=True)
```
